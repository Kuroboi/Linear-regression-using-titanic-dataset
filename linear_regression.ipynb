```python
import numpy as np
import pandas as pd
import math

trdata= pd.read_csv("titanic_train.csv")
tsdata=pd.read_csv("titanic_test.csv")
vdata=pd.read_csv("gender_submission.csv")

trdata.drop("Cabin",axis=1, inplace=True)
trdata["Age"].fillna(trdata["Age"].mean(skipna=True), inplace=True) #Missing values replaced by mean value of total age
trdata["Embarked"].fillna('Q', inplace=True) #Missing values(two) replaced by Q

tsdata.drop("Cabin",axis=1, inplace=True) 
tsdata["Age"].fillna(trdata["Age"].mean(skipna=True), inplace=True) #Missing values replaced by mean value of total age
tsdata['Fare'].fillna(trdata["Fare"].mean(skipna=True),inplace=True) #Missing values replaced by mean value of total Fare

sex=pd.get_dummies(trdata['Sex'],drop_first=True)
emb=pd.get_dummies(trdata['Embarked'], drop_first=True)
pcl=pd.get_dummies(trdata['Pclass'], drop_first=True)
trdata=pd.concat([trdata,sex,emb,pcl], axis=1)
trdata.drop(['Sex','Embarked','Pclass','Name','PassengerId','Ticket'],axis=1,inplace=True)

Name=tsdata['Name']
sex=pd.get_dummies(tsdata['Sex'],drop_first=True)
emb=pd.get_dummies(tsdata['Embarked'], drop_first=True)
pcl=pd.get_dummies(tsdata['Pclass'], drop_first=True)
tsdata=pd.concat([tsdata,sex,emb,pcl], axis=1)
tsdata.drop(['Sex','Embarked','Pclass','Name','PassengerId','Ticket'],axis=1,inplace=True)

X=trdata.drop(['Survived'],axis=1)
Y=trdata['Survived']
M=tsdata
N=vdata['Survived']

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,Y)
predictions=model.predict(M)
print("Passenger ID\t\tName\t\t\t\t\t\t\t\t\t\tSurvived?")
c=0
for i in range(len(predictions)):
    x=predictions[i]
    y=Name[i]
    if(x==0):
        print("{:<10}\t\t{:<70}\t\t{}".format(i,y,"NO"))
        c+=1
    else:
        print("{:<10}\t\t{:<70}\t\t{}".format(i,y,"YES"))
p=(len(M)-c)
print("\nTotal: \n Survived: {} \n Didnot survived: {}".format(p,c))

from sklearn.metrics import confusion_matrix as cf
print("Confusion Matrix:\n",cf(N,predictions))

#from sklearn.metrics import accuracy_score as ac
#print("\nAccuracy: ",ac(N,predictions))
```
